```python
converted_times = {}

# Convert 09:00:00 in each country's local time to Bangladesh time
for country, tz in time_zones.items():
    local_time = datetime.combine(datetime.now(), time(9, 0, 0))
    local_time = pytz.timezone(tz).localize(local_time)
    bd_time = local_time.astimezone(pytz.timezone('Asia/Singapore'))
    converted_times[country] = bd_time.strftime(' %H:%M')

# Create a DataFrame from the dictionary
df = pd.DataFrame(list(converted_times.items()), columns=['Country', 'Singapore Time'])

# Save the DataFrame to an Excel file
df.to_excel('converted_times_sing_min.xlsx', index=False)

print("Time conversion completed and saved to 'converted_times_singapore.xlsx'")
```
